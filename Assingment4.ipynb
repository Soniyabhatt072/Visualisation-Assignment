```python
#  5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
np.random.seed(42)
num_samples = 100
num_features = 5
data = pd.DataFrame(np.random.randn(num_samples, num_features))
data[1] = data[0] * 2 + np.random.normal(0, 0.5, num_samples)
data[2] = data[0] * -1 + np.random.normal(0, 0.3, num_samples)
data[3] = data[1] * 0.8 + np.random.normal(0, 0.4, num_samples)
data[4] = np.random.normal(0, 1, num_samples)
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f") # annot=True shows values, fmt=".2f" formats them
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", mask=mask)
plt.title("Correlation Matrix Heatmap (Upper Triangle)")
plt.show()
```

```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
# dimensional space.
# np.random.seed(30)
# data = {
#  'X': np.random.uniform(-10, 10, 300),
#  'Y': np.random.uniform(-10, 10, 300),
#  'Z': np.random.uniform(-10, 10, 300)
#  }
# df = pd.DataFrame(data)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(30)
data = {
 'X': np.random.uniform(-10, 10, 300),
 'Y': np.random.uniform(-10, 10, 300),
 'Z': np.random.uniform(-10, 10, 300)
 }
df = pd.DataFrame(data)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o', alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
# np.random.seed(15)
# data = {
#  'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
#  'Score': np.random.randint(50, 100, 200)
#  }
# df = pd.DataFrame(data)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(15)
data = {
 'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
 'Score': np.random.randint(50, 100, 200)
 }
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6)) # Adjust figure size for better visualization
sns.violinplot(x='Grade', y='Score', data=df, palette='Set3') # palette for colors
plt.title('Distribution of Scores Across Different Grades')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# np.random.seed(20)
# data = {
#  'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#  'Day': np.random.choice(range(1, 31), 100),
#  'Sales': np.random.randint(1000, 5000, 100)
#  }

# df = pd.DataFrame(data)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(20)
data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
 }

df = pd.DataFrame(data)
sales_matrix = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')
plt.figure(figsize=(10, 6))
sns.heatmap(sales_matrix, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
 }

df = pd.DataFrame(data)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8)) # Adjust figure size
ax = fig.add_subplot(111, projection='3d') # Create a 3D subplot

surface = ax.plot_surface(x, y, z, cmap='viridis') # Plot the surface
fig.colorbar(surface, shrink=0.75, pad=0.1) # Add a colorbar, shrink and pad are for better placement

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import bokeh.io
bokeh.io.output_notebook()
from bokeh.plotting import figure, show
import numpy as np
x = np.linspace(0, 10, 500)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 50, size=100)
colors = np.random.choice(['red', 'blue', 'green'], size=100)
p = figure(title="Bokeh Scatter Plot", x_axis_label="X-Axis", y_axis_label="Y-Axis")
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
#  fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#  counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruit=fruits, count=counts))
p = figure(x_range=fruits, height=350, title="Fruit Counts")
p.vbar(x='fruit', top='count', width=0.9, source=source)

p.title.align = "center"
p.xaxis.major_label_orientation = 0.8
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
#  data_hist = np.random.randn(1000)
#  hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.plotting import figure, show

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram of Random Data", x_axis_label="Value", y_axis_label="Frequency")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="black", alpha=0.7)
p.title.align = "center"
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import transform
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
source = ColumnDataSource(data=dict(x=xx.flatten(), y=yy.flatten(), z=data_heatmap.flatten()))
palette = "Viridis256"
color_mapper = LinearColorMapper(palette=palette, low=data_heatmap.min(), high=data_heatmap.max())
p = figure(width=400, height=400, title="Heatmap")
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=transform('z', color_mapper), line_color=None)
color_bar = ColorBar(color_mapper=color_mapper, ticker=p.xaxis[0].ticker,
                     label_standoff=12, border_line_color=None, location=(0,0))
p.xaxis.axis_label = "X"
p.yaxis.axis_label = "Y"
show(p)
```
